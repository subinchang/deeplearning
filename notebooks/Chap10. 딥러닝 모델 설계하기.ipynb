```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = np.loadtxt('../data/ThoraricSurgery.csv', delimiter=',')

print(type(dataset), dataset.shape, dataset.dtype)
```

```python
'''
문제1] 전체 환자의 진단 기록을 분리하여 X 로 저장하고, 생존 여부를 분리하여 Y 로 저장하는 코드를 작성하세요.
'''
```

```python
X = dataset[:,:17] # 1, 2차원 slicing
Y = dataset[:,17] # 1차원 slicking, 2차원 indexing

print(type(X), X.shape, X.dtype)
print(type(Y), Y.shape, Y.dtype)
```

```python
np.random.seed(1234)
tf.random.set_seed(1234)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=100, batch_size=10)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=17, activation='relu')) # 정해져 있음 (relu는 바꿀 수 있지만 권장X)
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 정해져 있음

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # 분류에선 accuracy가 가장 적절

model.fit(X, Y, epochs=500, batch_size=50) # epochs, batch_size 변경
```

```python
# 모델 epochs, 은닉층 수 늘려서 accuracy 높일 수 있지만 그게 더 좋은 모델이라고 할 수는 없음
# 이미 푼 참고서 여러 번 풀어서 정확도 높이는 것과 다를 바 없음
# test data에 대해서도 좋은 성능을 내야 좋은 모델
```
